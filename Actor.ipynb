```python
import testmodule
from datetime import datetime
import os
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
today_ = datetime.today().strftime('%Y-%m-%d')
# Folder containing CSV files
folder_path = './May 24/Daily files'

# Get list of all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Initialize an empty list to store DataFrames
data_frames = []

# Iterate over each CSV file, read it into a DataFrame, and append to the list
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    data_frames.append(df)

# Concatenate all DataFrames into one
combined_df = pd.concat(data_frames)

# Remove duplicates
combined_df.drop_duplicates(inplace=True)

# Save the combined and deduplicated DataFrame to a new CSV file
combined_df.to_csv('combined_data.csv', index=False)

print("Combined CSV file created successfully.")
```

```python
file = f'./combined_data.csv'

combined_df = pd.read_csv('combined_data.csv')
values = combined_df['Player Name']

value_counts = Counter(values)

# Get the top 20 most frequent values
top_20 = value_counts.most_common(40)

# Extract the values and counts from the top 20 items
value_labels, counts = zip(*top_20)

# Plot the counts against the values
plt.figure(figsize = (30,5))
plt.style.use('dark_background')
cmap = plt.get_cmap('Reds')
plt.bar(value_labels, counts)
plt.ylabel('Total deaths')
plt.xlabel('Player Name')
# plt.title('Top 20 Most Frequent Values')
plt.xticks(list(value_labels))
plt.xticks(rotation = 90, size = 25)
plt.yticks(rotation = 90, size = 15)
plt.xlim(-1, 40 - 0.5)
plt.savefig('./May 24/Daily photos/player_deaths', facecolor = 'Black', bbox_inches='tight')
plt.show()
```
